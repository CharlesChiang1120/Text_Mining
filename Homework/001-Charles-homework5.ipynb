## 請使用 jieba 對 article 的 content 進行切詞，並將切詞過後的資料放置於名為 corpus 的 list 中

```python
#article['content'] = article['content'].replace('\n',' ', regex=True) 
```

```python
corpus = []
jieba.add_word('一周')
jieba.add_word('狙擊')
jieba.add_word('擔心')
jieba.add_word('成長')
jieba.add_word('小時')
jieba.add_word('紐交所')
jieba.add_word('監管')
jieba.add_word('承擔')
jieba.add_word('持續')
jieba.add_word('前者')
jieba.add_word('玉山銀')
jieba.add_word('消費者')
jieba.add_word('不會')
jieba.add_word('新光三越')
jieba.add_word('引導')
jieba.add_word('藉此')
jieba.add_word('高盛')
jieba.add_word('元大證券')
jieba.add_word('老樂迪')
jieba.add_word('其他')
jieba.add_word('不是')
jieba.add_word('幾天')
jieba.add_word('續卡')
jieba.add_word('合併')
jieba.add_word('稅後')
jieba.add_word('每股盈餘')
jieba.add_word('中信')
jieba.add_word('股海')
jieba.add_word('明燈')
jieba.add_word('救濟金')
jieba.add_word('驚人')
jieba.add_word('紓困金')
jieba.add_word('成長')
jieba.add_word('折抵')
jieba.add_word('其實')
jieba.add_word('使用者')
jieba.add_word('個人')
jieba.add_word('不高興')
jieba.add_word('既視感')
jieba.add_word('這個')
jieba.add_word('炫晶卡')
jieba.add_word('信用卡')
jieba.add_word('一杯羹')
jieba.add_word('減產')
jieba.add_word('各種')
jieba.add_word('一億')
jieba.add_word('新戶')
jieba.add_word('不平衡')
jieba.add_word('稅後')
jieba.add_word('台壽')
jieba.add_word('操作')
jieba.add_word('存貨')
jieba.add_word('羥氯喹')
jieba.add_word('官網')
for content in article['content']:
    segwords=[]
    for word in jieba.cut(content):
        m =re.match('[\u4e00-\u9fa5]',word)
        if m:
            segwords.append(word)
    corpus.append(' '.join(segwords))
```

## 請使用 sklearn 將語料庫 Corpus 轉換成名為tfidf 的 TF-IDF矩陣 (TfidfVectorizer)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
tfidf  = vectorizer.fit_transform(corpus)
```

## 請使用 cosine distance 計算文章與文章間的距離，並將距離存到名為 cs 的變數中

```python
from sklearn.metrics.pairwise import cosine_distances
cs = cosine_distances(tfidf)
```

## 請使用 community detection 分群法將文章分群，並將分群後的結果存入 article_cluster 的變數名中

```python
#!pip3 install networkx
```

```python
import networkx as nx
m = (cs<=0.7).astype(int)
G = nx.from_numpy_matrix(m)
```

```python
#!pip3 install python-louvain
```

```python
import community
from collections import Counter
comm = community.best_partition(G)
c = Counter(comm.values())
```

```python
import numpy as np
title = article['title'].tolist()
np_title = np.array(title)
```

```python
from collections import Counter
article_cluster = {}
cluster_ary = np.array(list(comm.values()))
c = Counter(comm.values())
for group ,cnt in c.most_common(10):
    article_title =  np_title[cluster_ary == group]
    article_cluster[group] = article_title
```
