## 詞頻矩陣

```python
import jieba
jieba.add_word('台新')
jieba.add_word('富邦')
jieba.add_word('新戶')
jieba.add_word('本月')
jieba.add_word('振興券')
jieba.add_word('j卡')
jieba.add_word('三倍券')
jieba.add_word('不划算')
jieba.add_word('元大')
jieba.add_word('鑽金')
#jieba.load_userdict('./userdict.txt')
corpus=[]
for title in titles:
    corpus.append(' '.join(jieba.cut(title)))
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(stop_words=['不是'])
X = vectorizer.fit_transform(corpus)
```

```python
import re
corpus=[]
for title in titles:
    segwords=[]
    for word in jieba.cut(title):
        m =re.match('[\u4e00-\u9fa5]',word)
        if m:
            segwords.append(word)
    corpus.append(' '.join(segwords))
```

```python
import re
corpus = []
for title in titles:
    corpus.append(([word for word in jieba.cut(title) if re.match('[\u4e00-\u9fa5]+', word)]))
```

```python
flat_list = []
for sublist in corpus:
    for item in sublist:
        flat_list.append(item)
```

```python
flat_list
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(flat_list)
```

```python
print(vectorizer.get_feature_names())
```

```python
class ChineseCountVectorizer(CountVectorizer):
    def build_analyzer(self):
        analyzer = super(ChineseCountVectorizer, self).build_analyzer()
        return lambda doc: (w for w in analyzer(doc) if re.match('[\u4e00-\u9fa5]+', w))
```

```python
corpus=[]
for title in titles:
    corpus.append(" ".join(jieba.cut(title)))
```

```python
corpus
```

```python
vectorizer = ChineseCountVectorizer()
X = vectorizer.fit_transform(corpus)
```
