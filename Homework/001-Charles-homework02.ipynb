```python
import jieba
import pandas as pd
from collections import Counter
from wordcloud import WordCloud
```

```python
f = open('./article.txt','r')
article = f.read()
```

```python
article[0:100]
```

```python
article_split = list(jieba.cut(article))
```

```python
count = Counter(article_split)
```

```python
count.most_common(10)
```

```python
article_list = [i for i in list(jieba.cut(article))  if len(i)>=2]
```

```python
article_list[0:20]
```

```python
count = Counter(article_list)
```

```python
count.most_common(10)
```

```python
wc = WordCloud(background_color='black',width=500,height=600,font_path='./NotoSansCJKtc-Regular.otf')
wc.generate(' '.join(jieba.cut(article))) 
wc.to_image()
```
