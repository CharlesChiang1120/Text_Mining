```python
import pandas as pd
import numpy as np
```

```python
article = pd.read_excel('https://raw.githubusercontent.com/ywchiu/tibame_tm/master/data/finance_article.xlsx')
```

```python
article.head()
```

```python
import jieba
```

```python
corpus = []
titles = []
for idx, rec in article.iterrows():
    corpus.append(' '.join(jieba.cut(rec.get('content'))))
    titles.append(rec.get('title'))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(corpus)
```

```python
from sklearn.metrics.pairwise import cosine_distances
```

```python
cs = cosine_distances(tfidf)
```

```python
cs.shape
```

```python
cs
```

```python
m = (cs<0.5).astype(int)
```

```python
import networkx as nx
G = nx.from_numpy_matrix(m)
```

```python
import community
article_cluster = community.best_partition(G)
```
