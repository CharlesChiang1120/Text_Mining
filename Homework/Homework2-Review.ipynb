```python
with open ('./article.txt',encoding='utf-8') as f:
    data = f.read()
```

```python
list(jieba.cut(data))
```

```python
words = jieba.lcut(data)
```

```python
type(words)
```

## 方法一

```python
a = [1,1,2,2,1,3,1]
dic = {}
for w in a:
    if w not in dic:
        dic[w]=1
    else:
        dic[w]+=1
```

```python
dic
```

```python
dic={}
for word in words:
    if len(word)>=2:
        if word not in dic:
            dic[word]=1
        else:
            dic[word]+=1
    
```

```python
dic.items()
```

```python
swd = sorted(dic.items(),key = lambda e:e[1],reverse=True)
```

```python
swd[0:10]
```

## 方法二

```python
dic={}
for word in words:
    if len(word)>=2:
        dic[word]=dic.get(word,0) + 1 
```

```python
swd = sorted(dic.items(),key = lambda e:e[1],reverse=True)
swd[0:10]
```
